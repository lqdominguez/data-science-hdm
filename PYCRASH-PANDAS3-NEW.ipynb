```python
#PANDAS Data Frames
import pandas as pd

#to generate random numbers
from numpy.random import randn
```

```python
#create data frame size 5*5 containing random numbers
df = pd.DataFrame(randn(5,5),index=['ID1','ID2','ID3','ID4','ID5'],columns=['A1','A2','A3','A4','A5'])
df
```

# Join two data frames

```python
df2= pd.DataFrame(randn(3,3),index=['ID1','ID2','ID3'],columns=['A1','B2','B3'])
df2
```

```python
df.join(df2, lsuffix='_df1', rsuffix='_df2')
```

```python
#only use keys of right join partner
df.join(df2, lsuffix='_df', rsuffix='_df2', how='right')
```

# Append a data frame (using concat)

```python
#Append data frame 
#Append data frame to data frame
df_new = pd.DataFrame(randn(2,5),index=['ID6a','ID6b'],columns=df.columns)

#append command is deprecated
#df=df.append(df_new)
df = pd.concat([df, df_new])

df
```

```python
#Create a named series object
from numpy.random import rand

ser = pd.Series(rand(5),index=df.columns,name='ID6c')
ser
```

```python
#Append named series object to data frame
#df=df.append(ser)
series_df = ser.to_frame().T
df = pd.concat([df, series_df])

df
```

```python
#Append dictionary to data frame (without index)
dict_obj = {'A1': 0.1, 'A2': 0.2, 'A3': 0.3, 'A4': 0.4, 'A5': 0.5}
#df=df.append(dict_obj,ignore_index=True)

# Convert the dictionary to a DataFrame
dict_df = pd.DataFrame([dict_obj])

# Concatenate the DataFrame and the dictionary DataFrame vertically
df = pd.concat([df, dict_df])

df
```

```python
#Append named dictionary to data frame
dict_obj = {'A1': 0.2, 'A2': 0.4, 'A3': 0.6, 'A4': 0.8, 'A5': 1.0}

# Convert the dictionary to a DataFrame
dict_df = pd.DataFrame([dict_obj],index=['ID7'])

# Concatenate the DataFrame and the dictionary DataFrame vertically
df = pd.concat([df, dict_df])

df
```

```python
#Append list to data frame
#First try
new_list = [1,2,3,4,5]

# Convert the list to a DataFrame
list_df = pd.DataFrame([new_list])

# Concatenate the DataFrame and the list DataFrame vertically
df_temp = pd.concat([df, list_df])
df_temp
```

```python
#Append list to data frame
#Second try
# Convert the list to a DataFrame
list_df = pd.DataFrame([new_list], columns=df.columns)

# Concatenate the DataFrame and the list DataFrame vertically
df = pd.concat([df, list_df])
df
```
