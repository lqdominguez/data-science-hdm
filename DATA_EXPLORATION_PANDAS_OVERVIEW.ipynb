```python
import pandas as pd
from sklearn import datasets
```

# Load and access data in pandas data frame

```python
iris_df = datasets.load_iris(as_frame=True).frame
iris_df
```

```python
#show only selected columns (by name)
iris_df[['sepal width (cm)','target']]
```

```python
#show only selected columns (by index)
iris_df.iloc[:, 1:4]
```

```python
#select only selected columns (by index)
iris_df.iloc[5:10, :]
```

# Data Exploration: Calculate descriptive statistics

```python
#describe() shows most important descriptive statistics for all numeric columns of data frame
iris_df.describe()
```

```python
#Calculation further key figures about center of the data set
print("Mean of Sepal_Length",iris_df['sepal length (cm)'].mean())
print("Median of Sepal_Length",iris_df['sepal length (cm)'].median())
print("Mode of Sepal_Length",iris_df['sepal length (cm)'].mode())
```

```python
#Show correlations
corr = iris_df.corr()
corr
```

# Data Exploration: Visualization

```python
# Displaying correlations as a heatmap / correlation matrix 
corr.style.background_gradient(cmap ='coolwarm')
```

```python
# Scatter Plot
scatter = iris_df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', 
                           c='target',colormap='viridis',sharex=False)
```

```python
#Histograms of all dataframe features
hist = iris_df.iloc[:, 0:4].hist(bins=10,figsize=(10, 10))
```

```python
#Boxplots 
boxplot = iris_df.boxplot(figsize=(15, 10))
```
