# Load, explore and prepare data set

```python
import pandas as pd
import numpy as np

# Download housing dataset from UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
df = pd.read_csv(url, header=None, delim_whitespace=True)
df
```

```python
# Add column names to the dataframe
col_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS",
             "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
df.columns = col_names

# Add a nominal index feature
#df["suburb_index"] = range(1, len(df) + 1)

# Add two nominal condition features
df["condition"] = ["good" if i % 2 == 0 else "bad" for i in range(len(df))]
df["age"] = ["old" if i % 3 == 0 else "new" for i in range(len(df))]

df
```

```python
#seperate numeric and nominal features
numeric_features = df.select_dtypes(include=[np.number])
nominal_features = df.select_dtypes(include=[object])

#seperate label from numeric features
label = numeric_features.pop("MEDV")
```

# Feature selection by variance and variation ratio

Numeric features

```python
#min-max normalize numeric features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numeric_features = pd.DataFrame(scaler.fit_transform(numeric_features), columns=numeric_features.columns)
numeric_features.describe()
```

```python
#calculate variance of numeric features
numeric_features_var = numeric_features.var()
numeric_features_var
```

```python
#identify numeric features with low variance
numeric_features_low_var = numeric_features.loc[:, numeric_features.var() < 0.05]
numeric_features_low_var
```

Nominal features

```python
#describe nominal features
nominal_features.describe()
```

```python
#calculate variation ratio for nominal features
#explanation: The value_counts() method counts the number of occurrences of each unique value in the column, 
#and the normalize=True parameter normalizes the counts to represent percentages of the total number of values in the column
nominal_features_var = nominal_features.apply(lambda x: x.value_counts(normalize=True).max(), axis=0)
nominal_features_var_ratio=1-nominal_features_var
nominal_features_var_ratio
```

```python
#Alternatively, we could transform the nominal features to numeric ones
nominal_features_one_hot = pd.get_dummies(nominal_features)
nominal_features_one_hot
```

```python
#and calculate variance for them
nominal_features_one_hot_var = nominal_features_one_hot.var()
nominal_features_one_hot_var

#however these numbers could be misleading - why?
```

# Feature selection for numeric features using correlation and colinearity

```python
#caluclate correlation of numeric features with label
numeric_features_corr = numeric_features.corrwith(label)
numeric_features_corr
```

```python
#keep 10 best numeric features
numeric_features = numeric_features[numeric_features_corr.abs().sort_values(ascending=False).index[:10]]
numeric_features_corr = numeric_features.corrwith(label)
numeric_features_corr
```

```python
#calculate colinearity of remaining numeric features
numeric_features_corr = numeric_features.corr()
# Displaying correlations as a heatmap / correlation matrix
numeric_features_corr.style.background_gradient(cmap ='coolwarm')
```

```python
#apply a mask to highlight colinearity values > 0.5
numeric_features_corr_strong = (abs(numeric_features_corr) > 0.5)*1
# Displaying correlations as a heatmap / correlation matrix
numeric_features_corr_strong.style.background_gradient(cmap ='coolwarm')
```

```python
#calculate multicollinearity of numeric features
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(df):
    vif = pd.DataFrame()
    vif["features"] = df.columns
    vif["VIF Factor"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif
calc_vif(numeric_features)
```

```python
#remove TAX from numeric features
numeric_features = numeric_features.drop(columns=['TAX'])
#recalculate multicollinearity of numeric features#
calc_vif(numeric_features)
```

```python
#remove AGE from numeric features
numeric_features = numeric_features.drop(columns=['AGE'])
#recalculate multicollinearity of numeric features
calc_vif(numeric_features)
```

```python
#show remaining correlations with label
numeric_features_corr = numeric_features.corrwith(label)
numeric_features_corr
```

# Feature selection for nominal features using chi2 test

```python
# perform chi-squared test for nominal features
from scipy.stats import chi2_contingency

def calc_chi2(df):
    #df contains the two variables to be tested
    #----------step 2: Create a cross tab which is a contigency table
    data_cont=pd.crosstab(df.iloc[:,0], df.iloc[:,1])
    stat, p, dof, expected = chi2_contingency(data_cont)
    alpha = 0.05
    print("p value is " + str(p))
    if p <= alpha:
        print('Dependent (reject H0)')
    else:
        print('Independent (H0 holds true)')

#check independence of features to each other
calc_chi2(nominal_features)
```

```python
#check independence of features to label
nominal_label=pd.DataFrame()
#apply binning to label
nominal_label['label']=pd.cut(label, 2)

#first check label vs. label
nominal_label['label_copy']=pd.cut(label, 2)
print('Check independence of label and label')
calc_chi2(nominal_label.iloc[:,:2])

#check label vs. condition
nominal_label['condition']=nominal_features.iloc[:,0]
print('\nCheck independence of "condition" and label')
calc_chi2(nominal_label.loc[:,['label','condition']])

#check label vs. age
nominal_label['age']=nominal_features.iloc[:,1]
print('\nCheck independence of "age" and label')
calc_chi2(nominal_label.loc[:,['label','age']])
```
