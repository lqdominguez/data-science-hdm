# Univariate outlier detection

```python
#Calculate and print difference between mean and median for feature "distance" in the planets data frame
mean_median_difference_distance=planets_df['distance'].mean()-planets_df['distance'].median()
print(mean_median_difference_distance)

#Calculate and print skew for feature "distance" in the planets dataframe
skew_distance=planets_df['distance'].skew()
print(skew_distance)
```

```python
#Univariate outlier detection based on descriptive statistics (three standard deviations)

#calculate a variable which stores the value of three standard deviations of feature "distance"
three_std=planets_df['distance'].std()*3
#calculate a variable which stores the mean value of "distance"
mean=planets_df['distance'].mean()

#calculate threshhold min value for inliers (mean-three standard deviations)
inlier_low=mean-three_std
#calculate threshhold max value for inliers (mean-three standard deviations)
inlier_high=mean+three_std
#print thresshold min and max values
print("mean: ",mean, "lower boundary inlier: ",inlier_low, "upper boundary inlier",inlier_high,"\n")

#loop alle values in feature "distance", compare vales with thressholds and print if value is outlier
for value in planets_df['distance']:
    if value < inlier_low or value > inlier_high:
        print(value," is a potential outlier")    
```
