```python
#PANDAS Data Frames
import pandas as pd

#to generate random numbers
from numpy.random import randn
```

```python
#create data frame size 5*5 containing random numbers
df = pd.DataFrame(randn(5,5),index=['ID1','ID2','ID3','ID4','ID5'],columns=['A1','A2','A3','A4','A5'])
df
```

# Change column and index names

```python
#print columns and index
print(str(df.columns))
print(str(df.index))
```

```python
#change columns
df.columns=['B1', 'B2', 'B3', 'B4', 'B5']
df
```

```python
#change index
df.index=['First-ID', 'ID2', 'ID3', 'ID4', 'ID5']
#change columns
df.columns=['A1', 'A2', 'A3', 'A4', 'B5']
df
```

```python
#Change just one column
df=df.rename(columns = {'B5':'A5'})

#Change just one index
df=df.rename(index = {'First-ID':'ID1'})
df
```

# Selection and Slicing

```python
#Select values < 0
df_le1=df[df<0]
df_le1
#NaN=Not a Number
```

```python
#Drop NaN/missing values
df_le1.dropna()
```

```python
df
```

```python
# slice all rows, some columns 
df.loc[:, ['A2', 'A4']]
```

```python
# slice all columns, some rows
df.loc[['ID2', 'ID4'],:]
```

```python
# slicing by index: iloc
#first column and row: index = 0
#last index in range can equal, but not exceed number of rows

df.iloc[0:3, 1:4]
```

```python
#last single index in range needs to be smaller than number of rows
#this works
print(df.iloc[4, 0:4])

#this does not work
df.iloc[5, 0:4]
```

```python
# slice all columns, some rows
df.iloc[[2,4], :]
```

```python
# slice all rows, some columns
df.iloc[:, 2:4]
```

```python
# slice with range closed to one side
df.iloc[:, :3]
```

```python
#Combining positional and label-based indexing
#Single attribute
df.iloc[0: 3, df.columns.get_loc('A2')]
```

```python
#Combining positional and label-based indexing
#Multiple attributes
df.iloc[[0, 2], df.columns.get_indexer(['A3', 'A5'])]
```
